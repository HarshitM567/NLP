```python
import os

def merge_text_files(input_dir, output_file):
    with open(output_file, 'w') as output:
        for file_name in os.listdir(input_dir):
            if file_name.endswith('.txt'):  
                with open(os.path.join(input_dir, file_name), 'r') as input_file:
                    output.write(input_file.read())
                    output.write('\n')  
```

```python
def tokenization(file):
    num_tokens = 0
    with open(file, 'r', encoding='utf-8') as f:
        for line in f:
            tokens = word_tokenize(line)
            num_tokens += len(tokens)
    return os.path.basename(file), num_tokens


input_file = 'combined_text.txt'  

# Tokenize the file and print the file name and number of tokens
file_name, num_tokens = tokenization(input_file)
print(f"File: {file_name}, Number of tokens: {num_tokens}")
```

```python
def filterwords(file):
    with open(file, 'r',encoding = 'utf-8') as f:
        text = f.read()
        words = re.findall(r'\b\w+\b',text)
    ed_words = [word for word in words if re.search(r'\b\w+ed\b',word)]
    ing_words = [word for word in words if re.search(r'\b\w+ing\b', word)]
    es_words = [word for word in words if re.search(r'\b\w+es\b', word)]
    s_words = [word for word in words if re.search(r'\b\w+s\b', word)]
    un_words = [word for word in words if re.search(r'\bun\w+\b', word)]
    in_words = [word for word in words if re.search(r'\bin\w+\b', word)]
    
    return{"ed": ed_words,"ing": ing_words,"es": es_words,"s": s_words,"un": un_words,"in": in_words}

file = 'combined_text.txt'
extracted_words = filterwords(file)
for affix, words_list in extracted_words.items():
    print(f"Words with  {affix}  affix:{', '.join(words_list)}")
    
    
```

```python
def remove_stop_words(text):
    doc = nlp(text)
    return ' '.join(token.text for token in doc if token.text.lower() not in nlp.Defaults.stop_words)

file_path = "combined_text.txt"


num_words_before = 0


chunk_size = 10000  # Adjust this value as needed
with open(file_path, "r", encoding="utf-8") as file:
    while True:
        chunk = file.read(chunk_size)
        if not chunk:
            break
        num_words_before += len(chunk.split())
        processed_text = remove_stop_words(chunk)
        num_words_after = len(processed_text.split())

print("Number of words before stop word removal:", num_words_before)
print("Number of words after stop word removal:", num_words_after)
```

```python
def lemmatize_text(text):
    doc = nlp(text)
    return ' '.join(token.lemma_ for token in doc)

file_path = "combined_text.txt"

num_words_before = 0
chunk_size = 10000
with open(file_path, "r", encoding="utf-8") as file:
    while True:
        chunk = file.read(chunk_size)
        if not chunk:
            break
        num_words_before += len(chunk.split())
        lemmatized_text = lemmatize_text(chunk)
        num_words_before = len(lemmatized_text.split())


print("Number of words before lemmatization:", num_words_before)
print("Number of words after lemmatization:", num_words_after)
```

```python
#Harshit Mittal
#21BAI10058
#Q2
import re

def find_root_words(filename):
    with open(filename, 'r') as file:
        text = file.read()

    # Define a regular expression pattern to find words with silent 'e' before 'ing' and 'ed'
    pattern = re.compile(r'\b(\w+e)(ing|ed)\b')

    # Find all matches in the text
    matches = pattern.findall(text)

    # Extract root words from the matches
    root_words = [match[0] for match in matches]

    return root_words

# Replace 'shakes.txt' with the actual path to your file
filename = 'shakes.txt'
root_words = find_root_words(filename)

# Print the result
print("Root Words with silent 'e' dropped before 'ing' and 'ed':")
print(root_words)
```

```python
#21BAI10058 - Harshit Mittal
#For Any 10 Documents
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances, manhattan_distances, chebyshev_distances
from sklearn.metrics import jaccard_score
from scipy.spatial.distance import pdist, squareform
from scipy.stats import pearsonr
import numpy as np

# Load the text file
with open('bbc_news.txt', 'r', encoding='utf-8') as f:
    documents = f.readlines()

# Select the first 10 documents
documents = documents[:10]

# Initialize CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the documents
X = vectorizer.fit_transform(documents)

# Convert the matrix to array
X_array = X.toarray()

# Calculate Cosine Similarity
cos_sim = cosine_similarity(X_array)

# Calculate Euclidean Distance
euc_dist = euclidean_distances(X_array)

# Calculate Manhattan Distance
man_dist = manhattan_distances(X_array)

# Calculate Chebyshev Distance
che_dist = squareform(pdist(X_array, metric='chebyshev'))

# Calculate Jaccard's Coefficient
jac_score = jaccard_score(X_array[0], X_array[1], average='macro')

# Calculate Pearson's Coefficient
pearson_coef = pearsonr(X_array[0], X_array[1])

# Print the results
print("Cosine Similarity:\n", cos_sim)
print("Euclidean Distance:\n", euc_dist)
print("Manhattan Distance:\n", man_dist)
print("Chebyshev Distance:\n", che_dist)
print("Jaccard's Coefficient:\n", jac_score)
print("Pearson's Coefficient:\n", pearson_coef)
```

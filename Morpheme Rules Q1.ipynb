```python
#Harshit Mittal
#21BAI10058
#Q1
# Read the contents of "shakes.txt"
with open("shakes.txt", "r") as file:
    text = file.read()

# Split the text into words
words = text.split()

# Function to check if a word follows the specified rule
def follows_rule(word):
    vowels = "aeiou"
    if len(word) >= 5 and word[-3:] in ("ing", "ed"):  # Check if the word is long enough to contain the suffix
        if word[-4] not in vowels and word[-4] == word[-5]:  # Check consonant doubling
            return True
    return False

# Extract root words following the rule
root_words = []
for word in words:
    if follows_rule(word):
        root = word[:-4]  # Remove "ing" or "ed" to get the root form
        root_words.append((root, word))  # Store both root form and original form

# Remove duplicates
root_words = list(set(root_words))

# Print the root words along with their original forms
for root, original_form in root_words:
    print(root.upper() + "-" + original_form.upper())
```
